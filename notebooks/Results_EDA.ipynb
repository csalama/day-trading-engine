```python
import os
import pandas as pd
```

```python
PROJECT_PATH = os.path.dirname(os.getcwd())
DATASET_PATH = os.path.join(PROJECT_PATH,'data/processed/testing_data.csv')
OBSERVATION_PATH = os.path.join(PROJECT_PATH,'data/processed/MSFT_1year_feat.csv') 
NEW_DATASET_PATH = os.path.join(PROJECT_PATH,'data/processed/clean_testing_data.csv')
```

```python
def import_dataset(PATH):
    df = pd.read_csv(PATH)
    df['time'] = pd.to_datetime(df['time'])
    df.sort_values(by=['time'],inplace=True)
    df.dropna(inplace=True)
    df.reset_index(drop=True,inplace=True)
    return df
```

```python
df = pd.read_csv(DATASET_PATH)
```

```python
df
```

```python
#Concatenation did not exactly work properly... recreate the two dataframes
action_reward = df.loc[ (df['0'].isna())==True,:].dropna(axis=1).reset_index(drop=True)
```

```python
action_reward.info()
```

```python
observation = import_dataset(OBSERVATION_PATH) 
```

```python
observation = observation[0:20772].reset_index(drop=True )
```

```python
joint = action_reward.merge(observation,left_index=True,right_index=True)
joint
```

```python
joint['action_1'] = joint['action_tup'].str.replace(r'[^0-9\.\ +]','').str.split().str[0]
joint['action_2'] = joint['action_tup'].str.replace(r'[^0-9\.\ +]','').str.split().str[1]
```

```python
joint['action_1'] = joint['action_1'].str.replace(r'[\+]+','').astype(float)
joint['action_2'] = joint['action_2'].str.replace(r'[\+]+','').astype(float)
```

```python
joint['action_2']
```

```python
joint['action_str'] = 'hold'
joint.loc[ (joint['action_1'] < 2),'action_str' ] = 'sell'
joint.loc[ (joint['action_1'] < 1),'action_str'] = 'buy'
```

```python
joint['action_str'].value_counts()
```

```python
joint.to_csv(NEW_DATASET_PATH,index=False)
```
